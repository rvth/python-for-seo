```
!pip install pytrends
from pytrends.request import TrendReq
pytrends = TrendReq()
import pandas as pd                       
import time
import datetime
from datetime import datetime, date, time
```

Interest over Time


```
kw_list = ["labrador", "german shepherd", "staffordshire bull terrier", "rottweiler", "husky"]
pytrends.build_payload(kw_list, timeframe='today 3-m')
over_time = pytrends.interest_over_time()
over_time_df = pd.DataFrame(over_time)
over_time_df.drop(columns= 'isPartial') 
```

Related Queries

```
kw_list=['giant panda', 'tiger', 'polar bear', 'penguin']
pytrends.build_payload(kw_list, timeframe='today 3-m')
```

```
related_queries = pytrends.related_queries()
related_queries.values()
```

```
related_queries.get('giant panda').get('rising')
```

```
related_queries.get('tiger').get('top')
```

Suggestions

```
suggestions = pytrends.suggestions(keyword='nintendo switch')
suggestions_df = pd.DataFrame(suggestions)
suggestions_df.drop(columns= 'mid') 
```

Trending Searches


```
trend = pytrends.trending_searches(pn='united_kingdom')
uk_trend_df = pd.DataFrame(trend)
uk_trend_df
```

Interest by Region

```
pytrends.build_payload(kw_list=['coronavirus'], geo='GB') 
region_df = pytrends.interest_by_region(resolution='REGION', inc_low_vol=True)

region_df.head(10)
```

```
kw_list = ["BTS", "Stray Kids", "Day 6"]
pytrends.build_payload(kw_list)
kpop_regions_df = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True)

kpop_regions_df.head(10)
```
